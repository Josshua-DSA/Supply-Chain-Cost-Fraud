# AutoML untuk Shipment late Risk

```python
import pycaret
import pandas as pd

df = pd.read_csv("D:\PENS-EEPIS\SDT A Semester 4 2026\Project TA\dataset\DataCoSupplyChainDataset.csv", encoding='latin-1')
```

```python
# setup for classification
from pycaret.classification import *
setup = setup(data=df, target='Late_delivery_risk', session_id=123)
```

```python
# compare models
best_model = compare_models()
```

```python
# predict on test set
predict_model(best_model)
```

```python
# evaluate model
evaluate_model(best_model)
```

```python
# Plot Learning Curve
plot_model(best_model, plot='learning')
```
