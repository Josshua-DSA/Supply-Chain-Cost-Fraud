```python
# =========================
# 2. Parse date
# =========================

df = data_supply.copy()
df['order_date'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')
df['shipping_date'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')

# =========================
# 3. Date features 
# =========================
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_day'] = df['order_date'].dt.day
df['order_dayofweek'] = df['order_date'].dt.dayofweek
df['order_hour'] = df['order_date'].dt.hour
df['order_is_weekend'] = df['order_dayofweek'].isin([5, 6]).astype(int)


 # =========================
# 4. Shipping / delivery features
# Aman untuk fraud detection setelah transaksi selesai
 # =========================
df['actual_delay'] = (
        df['Days for shipping (real)'] -
        df['Days for shipment (scheduled)'])

df['is_late'] = (df['actual_delay'] > 0).astype(int)
df['is_early'] = (df['actual_delay'] < 0).astype(int)
df['severe_delay'] = (df['actual_delay'] >= 2).astype(int)

df['shipping_speed_ratio'] = (
        df['Days for shipping (real)'] /
        (df['Days for shipment (scheduled)'] + 1e-6)
    )

# =========================
# 5. Pricing / financial features
# =========================
df['expected_gross_sales'] = (
        df['Product Price'] * df['Order Item Quantity']
    )

df['price_gap'] = (
        df['Product Price'] -
        df['Order Item Product Price']
    )

df['price_gap_ratio'] = (
        df['price_gap'] /
        (df['Product Price'] + 1e-6)
    )

df['discount_to_sales_ratio'] = (
        df['Order Item Discount'] /
        (df['Sales'] + 1e-6)
    )

df['discount_to_gross_ratio'] = (
        df['Order Item Discount'] /
        (df['expected_gross_sales'] + 1e-6)
    )

df['sales_to_gross_ratio'] = (
        df['Sales'] /
        (df['expected_gross_sales'] + 1e-6)
    )

df['profit_margin'] = (
        df['Order Profit Per Order'] /
        (df['Sales'] + 1e-6)
    )

df['benefit_margin'] = (
        df['Benefit per order'] /
        (df['Sales'] + 1e-6)
    )

df['profit_per_item'] = (
        df['Order Profit Per Order'] /
        (df['Order Item Quantity'] + 1e-6)
    )

df['sales_per_item'] = (
        df['Sales'] /
        (df['Order Item Quantity'] + 1e-6)
    )

df['discount_per_item'] = (
        df['Order Item Discount'] /
        (df['Order Item Quantity'] + 1e-6)
    )
```

```python
# =========================
# 6. Consistency check features
# =========================
df['calculated_item_total'] = (
        df['Order Item Product Price'] *
        df['Order Item Quantity'] -
        df['Order Item Discount']
    )

df['item_total_gap'] = (
        df['Order Item Total'] -
        df['calculated_item_total']
    )

df['abs_item_total_gap'] = df['item_total_gap'].abs()

df['has_price_inconsistency'] = (
        df['abs_item_total_gap'] > 1e-3
    ).astype(int)

# =========================
# 7. Geo mismatch features
# =========================
df['country_mismatch'] = (
        df['Customer Country'] != df['Order Country']
    ).astype(int)

df['state_mismatch'] = (
        df['Customer State'] != df['Order State']
    ).astype(int)

df['city_mismatch'] = (
        df['Customer City'] != df['Order City']
    ).astype(int)

# =========================
# 8. Product / inventory simple features
# =========================
df['is_product_inactive'] = (
        df['Product Status'] == 0
    ).astype(int)

df['high_discount_flag'] = (
        df['Order Item Discount Rate'] > 0.5
    ).astype(int)

df['negative_profit_flag'] = (
        df['Order Profit Per Order'] < 0
    ).astype(int)

df['negative_margin_flag'] = (
        df['profit_margin'] < 0
    ).astype(int)

# =========================
# 9. Drop original date columns
# =========================
df = df.drop(
        columns=[
            'order date (DateOrders)',
            'shipping date (DateOrders)',
            'order_date',
            'shipping_date'
        ],
        errors='ignore'
    )
```
